```python
import torch
from torchvision import datasets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
%matplotlib inline


train_set = datasets.MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())
```

```python
# 显示图像
img_per_row = 8
fig = plt.figure(figsize=(10, 10), dpi=80)
for i in range(img_per_row * img_per_row):
    ax = fig.add_subplot(img_per_row, img_per_row, i+1)
    ax.imshow(train_set.data[i].numpy().reshape(28, 28), cmap=plt.cm.binary)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

```python
# 定义裁剪转换
c = transforms.CenterCrop((20, 20))
# 定义压缩转换
r = transforms.Resize((8, 8))
```

```python
def imshow(img):
    plt.imshow(img.squeeze(0).numpy(), cmap=plt.cm.binary)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
x = train_set[21][0]
imshow(x)
```

```python
# 将图片压缩到8x8，用于展示
imshow(r(c(x)))
torch.set_printoptions(precision=2)
print(r(c(x)))
```
