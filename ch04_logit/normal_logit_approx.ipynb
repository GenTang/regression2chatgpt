```python
from scipy.stats import norm, logistic
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 为在Matplotlib中显示中文，设置特殊字体
plt.rcParams['font.sans-serif'] = ['SimHei']
# 正确显示负号
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams.update({'font.size': 13})
# 创建一个图形框，在里面只画一幅图
fig = plt.figure(figsize=(9, 6), dpi=100)
ax = fig.add_subplot(111)
x = np.linspace(-8, 4, 100)
alpha = 1.702
normal = norm.cdf(x)
logit = logistic.cdf(alpha * x)
ax.plot(x, normal,
        label='标准正态分布：'
            + r'$F(x) = \int_{-\infty}^{x}\frac{1}{\sqrt{2\pi}}{\rm e}^{-\frac{t^{2}}{2}}$'
            + r'${\rm d}t$')
ax.plot(x, logit, 'k-.',
        label=r'最佳近似的逻辑分布：$F(x) = \frac{1}{1 + {\rm e}^{-1.702x}}$')
ax.set_xlabel('$x$')
ax.set_ylabel('累积分布函数：$F(x)$')
legend = plt.legend(shadow=True, fontsize=16)
plt.savefig('normal_logit_approx.png', dpi=200)
plt.show()
```
